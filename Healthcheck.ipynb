### Literally any first example from top of my head

```python
import keras
```

```python
import sklearn.datasets
```

```python
sklearn.datasets.load_iris().keys()
```

```python
X, Y = sklearn.datasets.load_iris()['data'], sklearn.datasets.load_iris()['target']
```

```python
from keras.layers import Dense
from keras.utils import to_categorical

m = keras.models.Sequential()
m.add(Dense(6, input_shape=(4,), activation='tanh'))
m.add(Dense(3, activation='softmax'))
m.compile('sgd', 'categorical_crossentropy')
m.fit(X,to_categorical(Y), epochs=1000)
```

```python
(m.predict(X).argmax(1) == Y).mean()
```
