```python
from fwk.stage import EqualLoudnessWeighting, Window, LogPowerFourier, PCENScaling
from fwk.dataset import Dataset
from fwk.acoustic import MappingGenerator

dset = Dataset()
dset.get_from("datasets/clarin-long/data")
dset.select_random(1)
```

```python
mapping = MappingGenerator([
    EqualLoudnessWeighting("A"),
    Window(512,128),
    LogPowerFourier()
])
mapping = mapping.get(dset)
dset.generate(mapping, ["stft", "clean"])
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
plt.figure(figsize=(20,10))
plt.imshow(-1 * np.flipud(dset.clean[0].T), aspect='auto')
```

```python
plt.figure(figsize=(20,10))
plt.imshow(np.flipud(dset.stft[0].T), aspect='auto')
```
